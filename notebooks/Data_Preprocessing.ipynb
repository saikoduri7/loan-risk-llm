```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('accepted_2007_to_2018Q4.csv', low_memory=False)
```

```python
df.head()
```

```python
for i, col in enumerate(df.columns):
  print(f'{i}: {col}')
```

```python
df = df[['loan_amnt', 'term', 'int_rate', 'grade', 'emp_length', 'home_ownership',
    'annual_inc', 'purpose', 'addr_state', 'dti', 'delinq_2yrs',
    'fico_range_high', 'loan_status']] #keeping only relevant columns
```

```python
df.head()
```

```python
df = df.dropna() #dropping all missing values
```

```python
df
```

```python
df['loan_status'].isin(['Fully Paid', 'Charged Off'])
```

```python
df = df[df['loan_status'].isin(['Fully Paid', 'Charged Off'])]
#filters to only loans that were either fully paid or charged off
```

```python
df
```

```python
df['risk'] = df['loan_status'].apply(lambda x: 'LOW RISK' if x == 'Fully Paid' else 'HIGH RISK')
```

```python
df.head()
```

```python
df[['loan_amnt','annual_inc','fico_range_high','emp_length','purpose','risk']]
```

```python
def format_row(row):
  prompt = (f"Loan Amount: ${int(row['loan_amnt'])}",f"Annual Income: ${int(row['annual_inc'])}",
            f"FICO Score: {int(row['fico_range_high'])}",f"Employment Length: {row['emp_length']}",
            f"Purpose: {row['purpose']}")
  return {"input": prompt, "output": row['risk']}

formatted_data = df.apply(format_row, axis=1)
```

```python
formatted_data[0] #now all rows are formatted
```

```python
len(formatted_data)
```

```python
import json  #need to save as json to fine tune
```

```python
with open("loan_risk_data.json", "w") as f:
  json.dump(formatted_data.tolist(), f, indent=2)

print("saved as json")
```

```python
with open('loan_risk_data.json', 'r') as f:
    data = json.load(f)
```

```python
with open('loan_risk_data.jsonl', 'w') as out:
    for item in data:
        prompt = ", ".join(item["input"]) + ". What is the risk level?"
        response = item["output"]
        json.dump({"prompt": prompt, "response": response}, out)
        out.write('\n')
```
